```python
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2) # this will generate 1 x 2
ax[0].plot(x, cos_x, label='cos x')
ax[0].legend()

ax[1].plot(x, sin_x, label='sin x')
ax[1].legend()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1) # this will generate 2x1 

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()

type(ax) #numpy.ndarray
ax.shape #(2,) size
type(ax[0]) #this is a matplotlib axis 
```

A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```

plt plot vs ax.plot? 
- ax.plot allows modifications to be made more easily

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
ax[0].plot(x, cos_x, label='cos x', c='r',ls='--')
ax[0].legend()
ax[0].grid(alpha=0.25) #this is repeated

ax[1].plot(x, sin_x, label='sin x')
ax[1].legend()
ax[1].grid(alpha=0.25) #this is repeated

ax[0].set_xlabel('Hahaha')
```

Understanding plot creation 

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```
