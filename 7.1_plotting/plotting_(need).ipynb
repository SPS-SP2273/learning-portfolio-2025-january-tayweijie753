```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]


# Let's start plotting
#plt.plot(x, y_1, label='Y values',#plt.plot(x,y) itself will print the graph; the others are for design/colors/labels
#        color='red', linestyle='dashed')

plt.scatter(x, y_1, label='Y1 values', ls="dashed",
            c='maroon')
plt.scatter(x, y_2, label='Y2 values',
           c= 'b')
plt.xlabel('x-values') #label x axis
plt.ylabel('y-values') #label y axis
plt.title('X vs Y')
plt.grid(alpha=.25) #controls how opaque/transparent the grid is - the higher value = more bold lines 
plt.legend(loc='upper left') #loc is location - canchange to right/left, or a coordinate 
plt.savefig("xvsy.png") #this is used to save image as png, jpg and PDF 
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()   
```
