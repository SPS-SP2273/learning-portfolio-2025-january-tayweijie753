## Exercise 3 (Timing Stuff)

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

The first function processes output faster than the second function (252ns vs 285ns) after running a total of 7,000,000 loops. 
1 run is 1,000,000 loops. A total of 7 runs was runned, leading to a total of 7,000,000 loops runned. 
The output timing (252+/-sd and 282ns+/-sd) is the mean+SD of the all 7,000,000 loops runned.
